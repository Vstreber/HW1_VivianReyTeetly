```python
import pandas as pd
csv = pd.read_csv('candy-data.csv')
```

```python
csv
```

```python
#convert winpercent to decimal
csv['winpercent'] = csv['winpercent'] / 100
#mse error 
from sklearn.metrics import mean_squared_error
mse_win_percent = mean_squared_error(csv['winpercent'], csv['win-prediction'])
print("win percent mse error: ", mse_win_percent)
#mse error chocolate prdiction
mse_chocolate = mean_squared_error(csv['chocolate'], csv['chocolate-prediction'])
print("chocolate mse error: ", mse_chocolate)
```
